## Descrição da Base de Dados:

  Vídeos diários de tendências do YouTube. A seção "Em Alta" do YouTube tem por objetivo ajudar os espectadores a ver o que está acontecendo no YouTube e no mundo destacando vídeos que podem interessar a vários espectadores. Os vídeos são considerados "em alta" devido a combinação dos seguintes fatores: 
  
      * Número de vizualizações;
      * Velocidade com que o vídeo gera visualizações;
      * Origem das visualizações;
      * A quanto tempo o vídeo foi enviado.
      
  O conjunto de dados inclui dados do período de 14/11/2017 até 14/06/2018. Os dados são das regiões dos Estados Unidos, Grã-Bretanha, Alemanha, Canadá, França, Rússia, México, Coréia do Sul, Japão e Índia.

## Leitura dos dados:

```r
library(repr)
library(readr)

USvideos <- read_csv('C:\\Users\\jotap\\Documents\\Data\\youtube-new\\Nova pasta\\USvideos.csv')
RUvideos <- read_csv('C:\\Users\\jotap\\Documents\\Data\\youtube-new\\Nova pasta\\RUvideos.csv')


head(USvideos)
head(RUvideos)
options(warn= -1)
```

 ## Limpeza dos dados:
 
 Primeiramente, usamos o pacote lubridate para fazer o R converter e entender as colunas de data como datas:

```r
library(lubridate)
#US
USvideos$trending_date <- ydm(USvideos$trending_date)
USvideos$publish_time <- ymd_hms(USvideos$publish_time)

#Rússia
RUvideos$trending_date <- ydm(RUvideos$trending_date)
RUvideos$publish_time <- ymd_hms(RUvideos$publish_time)
```


Removemos as colunas que não julgamos interessantes para fazer análises: _thumbnail_link_, _video_error_or_removed_ e _description_, e transformamos o nome dos canais em 'factor'.


```r
USvideos <- USvideos[, - c(12, 15, 16)]
USvideos$channel_title <- as.factor(USvideos$channel_title)

RUvideos <- RUvideos[, - c(12, 15, 16)]
RUvideos$channel_title <- as.factor(RUvideos$channel_title)
```

---
## Extração de categorias:

As categorias estão em números, mas não é interessante analisar sem saber do que se trata, então extraímos as categorias do arquivo .json que veio junto.


```r
library(RJSONIO)

category_genre <- fromJSON("C:\\Users\\jotap\\Documents\\Data\\youtube-new\\Nova pasta\\US_category_id.json")

category <- c(NULL)
category_id <- c(NULL)
for(i in 1:length(category_genre$items)){
  category <- append(category, category_genre$items[i][[1]]$snippet$title)
  category_id <- append(category_id, category_genre$items[i][[1]]$id)
}

df <- data.frame(cbind(category, category_id))
```

---
## Inserindo as categorias as bases:


```r
#EUA:
final_df_US <- merge(USvideos, df, by='category_id', all.x = TRUE)
final_df_US <- final_df_US[,- c(1)]
final_df_US$category  <- factor(final_df_US$category)

#RÚSSIA:
final_df_RU <- merge(RUvideos, df, by='category_id', all.x = TRUE)
final_df_RU <- final_df_RU[,- c(1)]
final_df_RU$category  <- factor(final_df_RU$category)
```

## Modelo

### O modelo que escolhemos foi o Random Forest para prever de que categoria novos videos que fossem inseridos no dataset seriam baseado em suas características

```r
US_model <- subset(final_df_US, likes>0 & comment_count>0)
```

#### Dividimos o dataset em treino e teste contendo, respectivamente, 80 e 20% dos dados

### Assim poderemos treinar o modelo em dados que ele nunca "viu" e avaliar sua performance

```r
library(caret)
set.seed(1)

training_set <- createDataPartition(y = US_model$category, p = 0.8, list = FALSE)
train_US <- US_model[training_set,]
test_US <- US_model[-training_set,]
```

```r
set.seed(1)
randfor <- train(category ~ views + tag_num + engagement + time_until_viral, 
             data = train_US, method = "rf")
```

```r
pred <- predict(randfor, test_US)
confusionMatrix(pred, test_US$category)
```

### O modelo de Random Forest para os EUA teve uma acurácia de 67.66%, baseado nas features 'views', numero de tags, engajamento e tempo até se tornar viral

### Agora faremos o mesmo para a Rússia

```r
RU_model <- subset(final_df_RU, likes>0 & comment_count>0)
```

```r
training_set_ru <- createDataPartition(y = RU_model$category, p = 0.8, list = FALSE)
train_RU <- RU_model[training_set_ru,]
test_RU <- RU_model[-training_set_ru,]
```

```r
set.seed(1)
ru_randfor <- train(category ~ views + engagement + tag_num + time_until_viral, 
             data = train_RU, method = "rf")
```

### O modelo de Random Forest para a Rússia teve uma acurácia de 28%, baseado nas features 'views', numero de tags e tempo até se tornar viral

```r
pred_ru <- predict(ru_randfor, test_RU)
confusionMatrix(pred_ru, test_RU$category)
```
